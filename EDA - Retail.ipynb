```python
print(data['Profit'].quantile(0.10))
print(data['Profit'].quantile(0.90))
```

```python
data['Profit'] = np.where(data['Profit']<-15.47346,-15.47346,data['Profit'])
data['Profit'] = np.where(data['Profit']>89.28162,89.28162,data['Profit'])
print(data['Profit'].skew())
```

# Lineplots

```python
plt.figure(figsize=(12,7))
sns.lineplot('Quantity','Profit',data=data,color='blue',label='Profit VS Quantity')
```

 We can see with increase in Quantity , the profit also increases.

```python
plt.figure(figsize=(12,7))
sns.lineplot('Discount','Profit',data=data,color='red',label='Profit VS Discount')
```

We can see here that with increase in Discount , the Profit decreases thats understandable.
